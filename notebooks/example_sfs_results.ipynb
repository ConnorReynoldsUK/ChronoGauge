# Interpreting the results of the sequential feature selection (SFS) wrapper
This notebook gives our method for interpreting the SFS results. Here, we take all outputs across all iterations of each SFS run and find the gene set which gave the lowest mean-absolute-error (MAE).

```python
import pandas as pd
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt
```

First, we read each the pickle output of each iteration, which includes the cross-validation MAE and the gene feature set used in said iteration.

```python
results_errs = {}
results_genes = {}
for r,d,f in os.walk('../results/sfs_results/seed_0_103613_260924//feature_sets'):
    for file in f:
        with open(r+'/'+file, 'rb') as fin:
            i_results = pickle.load(fin)
            results_errs[i_results['iteration']] = i_results['error']
            results_genes[i_results['iteration']] = i_results['gene_set']
```

Select the iteration giving the smallest MAE

```python
# get minimum mean-absolute-error across feature sets
minimum_index = np.argmin(list(results_errs.values()))
iter_min = list(results_errs.keys())[minimum_index]
err_min = list(results_errs.values())[minimum_index]
genes_min = list(results_genes.values())[minimum_index]
print('Minimum 5-fold cross-validation mean-absolute-error: {:.2f} mins at iteration {} with a gene length of {}'.format(err_min, iter_min, len(genes_min)))
```

We find that the gene set in the final iteration (49) gave the smallest MAE. This set can be used to tune and train an individual sub-predictor forming part of ChronoGauge's ensemble.

```python
# sort by iteration
sorted_errs = dict(sorted(results_errs.items()))

plt.plot(list(sorted_errs.keys()), list(sorted_errs.values()))
plt.scatter(iter_min, err_min, c='r')
plt.xlabel('Sequential feature selection iteration')
plt.ylabel('5-fold cross-validation\nmean-absolute-error (mins)')
plt.savefig('sfs_example.png')
```
